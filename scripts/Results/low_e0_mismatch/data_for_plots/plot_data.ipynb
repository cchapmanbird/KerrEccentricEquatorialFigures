```python
import matplotlib.pyplot as plt
import numpy as np
from seaborn import color_palette

plt.rcParams["text.usetex"] = True
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Computer Modern"]
cpal = color_palette("colorblind", 4)

SNR_choice = np.load("SNR_choice.npy")
little_e0 = np.load("e0_vec.npy")
mismatch_vec_np = np.load("mismatch_vec.npy")

e0_fit = np.load("e0_for_fit.npy")
mismatch_fit = np.load("mismatch_fit.npy")

plt.figure(figsize=(5,4))
plt.loglog(little_e0, mismatch_vec_np, 'o', ms = 6, color=cpal[0])
plt.loglog(e0_fit, mismatch_fit, linestyle = 'dashed', c = 'red', label = r"$\propto (e_{0})^{4}$")

plt.axhline(y = (12/(2*100**2)), c = 'black', linestyle = 'dashed', label = r'Lindblom Criterion: $D/(2\rho^{2})$')
plt.xlabel(r'$e_{0}$', fontsize = 18)
plt.ylabel(r'$\mathcal{M}(h^\mathrm{circ}, h^\mathrm{ecc})$', fontsize = 18)
# plt.title(r'$(M,\mu,a,SNR,T_{obs}) = (10^{6}M_{\odot}, 25M_{\odot}, 0.998, 50, 2 \, yr)$', fontsize = 12)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.legend(fontsize = 13, frameon=False)
plt.ylim(1e-11, 5e-1)
plt.savefig("./plots/low_e0_mismatch_ODE_1e-11_gpu_tdi2_w_background_paper_run.pdf",bbox_inches = "tight")
```
