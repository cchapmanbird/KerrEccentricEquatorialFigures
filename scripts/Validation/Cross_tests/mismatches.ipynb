```python
import json
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
data_path = "./Data/mismatch_across_models_full_newmass.json"
with open(data_path, "r") as f:
    data = json.load(f)
```

```python
data["a0:-0.09"]
```

```python
a_keys = data.keys()

a0_arr = np.zeros(len(a_keys))
mis_arr = np.zeros((len(a_keys),2))
for ii,a_key in enumerate(a_keys):
    a0_arr[ii] =  data[a_key]["a0"]
    mis_arr[ii,0] = data[a_key]["mismatch_KerrCirc_l10_vs_KerrEcc"]
    mis_arr[ii,1] = data[a_key]["mismatch_BHPWave_vs_KerrEcc"]
    # print(a0_arr[ii], mis_arr[ii,0], mis_arr[ii,1])



plt.figure(figsize=(8,6))
plt.plot(a0_arr, mis_arr[:,0], label="KerrCirc vs Ours")
plt.plot(a0_arr, mis_arr[:,1], label="BHPWave vs Ours")

plt.xlabel("Primary Spin $(a)$")
plt.ylabel(r"Mismatch $\mathcal{M}$")
plt.yscale("log")
plt.grid()
plt.legend()
```
