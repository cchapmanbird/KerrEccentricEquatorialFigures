```python
data_path = "./Data/mismatch_across_models_full_newmass.json"
with open(data_path, "r") as f:
    data = json.load(f)



a_keys = data.keys()

a0_arr = np.zeros(len(a_keys))
mis_arr = np.zeros((len(a_keys),2))
for ii,a_key in enumerate(a_keys):
    a0_arr[ii] =  data[a_key]["a0"]
    mis_arr[ii,0] = data[a_key]["mismatch_KerrCirc_l10_vs_KerrEcc"]
    mis_arr[ii,1] = data[a_key]["mismatch_BHPWave_vs_KerrEcc"]
    # print(a0_arr[ii], mis_arr[ii,0], mis_arr[ii,1])



# data_path_2 = "./Data/mismatch_across_models_full_newmass_lamx9BHPWave.json"
# with open(data_path_2, "r") as f:
#     data_2 = json.load(f)

# a_keys_2 = data_2.keys()
# a0_arr_2 = np.zeros(len(a_keys_2))
# mis_arr_2 = np.zeros((len(a_keys_2),3))
# for ii,a_key in enumerate(a_keys_2):
#     a0_arr_2[ii] =  data_2[a_key]["a0"]
#     mis_arr_2[ii,0] = data_2[a_key]["mismatch_KerrCirc_l10_vs_KerrEcc"]
#     mis_arr_2[ii,1] = data_2[a_key]["mismatch_BHPWave_vs_KerrEcc"]
#     mis_arr_2[ii,2] = data_2[a_key]["mismatch_KerrCirc_l10_vs_BHPWave"]
```

```python

title_fontsize = 20
label_fontsize = 16
tick_fontsize = 16
legend_fontsize = 14
text_fontsize = 14


plt.figure(figsize=(7,4.5))
plt.plot(a0_arr, mis_arr[:,0],'-*k', label="KerrCirc vs Ours",rasterized=True)
plt.plot(a0_arr, mis_arr[:,1],'-*r', label="BHPWave vs Ours", rasterized=True)

# plt.plot(a0_arr_2, mis_arr_2[:,0],'--*b', label="KerrCirc vs Ours (l=10)", rasterized=True)
# plt.plot(a0_arr_2, mis_arr_2[:,1],'--*g', label="BHPWave vs Ours (l=10)", rasterized=True)
# plt.plot(a0_arr_2, mis_arr_2[:,2],'--*m', label="KerrCirc vs BHPWave (l=10)", rasterized=True)

plt.xlabel("Primary Spin $(a)$", fontsize=label_fontsize)
plt.ylabel(r"Mismatch $\mathcal{M}$", fontsize=label_fontsize)
plt.yscale("log")
plt.tick_params(axis='both', which='major', labelsize=tick_fontsize)
# plt.grid()
plt.legend(loc="upper left", fontsize=legend_fontsize)
plt.savefig("mismatch_across_models_full.pdf", bbox_inches="tight")
```
